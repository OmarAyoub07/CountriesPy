```python
# Block 1: Import necessary libraries and define helper functions
import requests
import pandas as pd
from IPython.display import display


# Helper function to make API requests and handle errors
def fetch_data(url):
    try:
        response = requests.get(url)
        response.raise_for_status()  # Raise an error for bad status codes
        return response.json()
    except requests.RequestException as e:
        print(f"Error fetching data: {e}")
        return None
```

```python
from deep_translator import GoogleTranslator

def translate_text(text, target_language):
    translator = GoogleTranslator(source='auto', target=target_language)
    translated = translator.translate(text)
    return translated

text_to_translate = "Hello, world!"
target_language = "ar"

translated_text = translate_text(text_to_translate, target_language)
print(translated_text)
```

```python
# countries end-point
url = "https://restcountries.com/v3.1/all?fields=name,translations,cca2,idd"

try:
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        print("Successfully retrieved")

        for country in data:
            print(f'''
                Country Code:    {country["cca2"]}
                Country Name En: {country["name"]["common"]}
                Country Name Ar: {country["translations"]["ara"]["common"]}
                Country Phone:   {country["idd"]["root"] + country["idd"]["suffixes"][0] if country["idd"].get("suffixes") else country["idd"]["root"]}''')
    else:
        print("failed")
except requests.exceptions.RequestException as e:
    print(e)
```

```python
# Block 3: Fetch regions based on selected country (using a placeholder API)
# Replace 'SA' with the country code of your choice
lang = "ar"
selected_country = "SA"
regions_url = f"https://countriesnow.space/api/v0.1/countries/states"
regions_data = fetch_data(regions_url)

if regions_data and "data" in regions_data:
    regions = next((c for c in regions_data["data"] if selected_country in c["iso2"]), None)
    regions_df = pd.DataFrame(regions["states"], columns=["name", "state_code"])
    if lang != "en":
        regions_df["name"] = [translate_text(c, "ar") for c in regions_df["name"]]
    display(regions_df)
else:
    print("Failed to fetch regions data.")
```

```python
country = regions["name"]
region = regions["states"][8]["name"]
cities_url = "https://countriesnow.space/api/v0.1/countries/state/cities"
body = {
    "country": country,
    "state": region
}

response = requests.post(cities_url, json = body)

if response.status_code == 200:
    cities = response.json()
    cities_df = pd.DataFrame(cities["data"], columns=["City Name"])
    if lang != "en":
        cities_df["City Name"] = [translate_text(c, "ar") for c in cities_df["City Name"]]
    display(cities_df)
```
